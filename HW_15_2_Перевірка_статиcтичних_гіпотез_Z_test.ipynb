```python
import numpy as np
from scipy import stats
```

```python
sample_mean = np.mean(data)
population_mean = 20
population_std = np.std(data, ddof=1)
sample_size = len(data)
alpha = 0.05
```

Гіпотези

H₀: μ = 20 (середній чек залишився $20)

H₁: μ > 20 (середній чек підвищився)

Тип тесту: правосторонній (one-tailed)

Чому саме правосторонній тест?

Ми маємо перевірити, чи вдалося саме підвищити середній чек за липень 2025, а не просто сума в чеку відрізняється від історичного середнього.

```python
# Формула: z = (x̄ - μ₀) / SE
z_stat = (sample_mean - population_mean) / SE

print(f"z = {z_stat:.3f}")
```

```python
# Для правостороннього тесту з α = 0.05
z_crit = stats.norm.ppf(1 - alpha)

print(f"Критичне z-значення = {z_crit:.3f}")
```

```python
# Для правостороннього тесту: p-value = P(Z > z_stat)
p_value = 1 - stats.norm.cdf(z_stat)

print(f"p-value = {p_value:.6f}")
```

```python
print("=" * 60)
print("МЕТОД 1: Порівняння z-статистики з критичним значенням")
print("=" * 60)
print(f"z-статистика = {z_stat:.3f}")
print(f"Критичне значення = {z_crit:.3f}")
print(f"Порівняння: {z_stat:.3f} {'>' if z_stat > z_crit else '≤'} {z_crit:.3f}")

if z_stat > z_crit:
    print("Висновок: z-статистика ПЕРЕВИЩУЄ критичне значення")
    decision1 = "ВІДХИЛЯЄМО H₀"
else:
    print("Висновок: z-статистика НЕ перевищує критичне значення")
    decision1 = "НЕ ВІДХИЛЯЄМО H₀"

print(f"Рішення: {decision1}")
```

Чи зійшлись значення z-статистику та р-значення в цьому завданні з попередніми обчисленнями?

```python
from statsmodels.stats.weightstats import ztest
z_stat, p_value = ztest(data, value=20, alternative='larger')

print(f"  Z-статистика: {z_stat:.3f}")
print(f"  P-value: {p_value:.6f}")
```

```python
alpha = 0.05

if p_value < alpha:
    print("Відхиляємо H0: середній чек статистично значуще зріс")
else:
    print("Не відхиляємо H₀: недостатньо доказів")
```

За результатами z-тесту отримано p-value. Оскільки p-value > 0.05, ми не відхиляємо нульову гіпотезу. Немає статистично значущих підстав вважати, що середній чек зріс.
